```python
#Q8 produce a list of booking on the day of 2012-09-14, which will cost memeber or gust more than $30
# remember that memeber and guest have different price. The cost are half hour per slot and guest userid is usually 0
# output: name of facilities, memeber formatted in a single column, the descending order of cost, do not use subquery
import sqlite3
from sqlite3 import Error
from pandas import DataFrame

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    query1 = """
    SELECT m.firstname || ' ' || m.surname AS fullname, b.starttime,
    CASE WHEN m.memid = 0 THEN (f.guestcost * b.slots)
    ELSE (f.membercost * b.slots)
    END AS cost, f.name
    FROM Members m
    LEFT JOIN Bookings b ON m.memid = b.memid
    LEFT JOIN Facilities f ON b.facid = f.facid
    WHERE b.starttime LIKE '2012-09-14%'
    AND IIF(m.memid =0, f.guestcost, f.membercost) * b.slots > 30
    ORDER BY cost DESC    
        """
    cur.execute(query1)
 
    rows = cur.fetchall()

    for row in rows:
        print(row)


def main():
    database = "C:\\Users\\mohua\\OneDrive\\Documents\\unit6\\sqlite\\db\\pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
#Q10 product a list of facility where the total revenue was less than 1000. 
# remember that memeber and guest have different price. The cost are half hour per slot and guest userid is usually 0
# output: name of facilities, total revenue on descending order
import sqlite3
from sqlite3 import Error
from pandas import DataFrame

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    query1 = """
    SELECT name, revenue
    FROM (
    SELECT name, SUM(
    CASE WHEN memid =0
    THEN guestcost * slots
    ELSE membercost * slots
    END ) AS revenue
    FROM Bookings b
    INNER JOIN Facilities f ON b.facid = f.facid
    GROUP BY name
    ) AS inner_table
    WHERE revenue <1000
    ORDER BY revenue
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
#    df = DataFrame(rows,columns=['Tennis Court','firstname','surname'])
#    print(df)
    for row in rows:
        print(row)


def main():
    database = "C:\\Users\\mohua\\OneDrive\\Documents\\unit6\\sqlite\\db\\pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
